```python
def get_name(row):
    print('here')
    writers = row.writer.split(',')
    directors = row.director.split(',')
    print(writers)
    for i in range(len(writers)):
        writers[i] = names[names.nconst == writers[i]].primaryName.to_list()[0]
    for i in range(len(directors)):
        directors[i] = names[names.nconst == directors[i]].primaryName.to_list()[0]
    print(directors)
    row.writer = ','.join(writers)
    row.director= ','.join(directors)
    print(row)
    return row
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['world_revenue'], axis=1).values
y = df.world_revenue.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# pip install xgboost
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
import numpy as np
```

```python
model = RandomForestRegressor(max_features= 'auto',min_samples_split=6,min_impurity_decrease=0.25,min_samples_leaf=4,n_jobs=-1,random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
round(model.score(X_train, y_train), 2) * 100
```

```python
mean_squared_log_error(model.predict(X_train), y_train)**0.5
```

```python
round(model.score(X_test, y_test), 2) * 100
```

```python
mean_squared_log_error(model.predict(X_test), y_test)**0.5
```

```python
mean_squared_error(model.predict(X_test),y_test)**0.5
```
